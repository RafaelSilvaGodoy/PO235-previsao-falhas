```python
import numpy as np
import pandas as pd
```

```python
index_names = ['id', 'time_cycles']
setting_names = ['setting_1', 'setting_2', 'setting_3']
sensor_names = ["(Fan inlet temperature) (◦R)",
    "(LPC outlet temperature) (◦R)",
    "(HPC outlet temperature) (◦R)",
    "(LPT outlet temperature) (◦R)",
    "(Fan inlet Pressure) (psia)",
    "(bypass-duct pressure) (psia)",
    "(HPC outlet pressure) (psia)",
    "(Physical fan speed) (rpm)",
    "(Physical core speed) (rpm)",
    "(Engine pressure ratio(P50/P2)",
    "(HPC outlet Static pressure) (psia)",
    "(Ratio of fuel flow to Ps30) (pps/psia)",
    "(Corrected fan speed) (rpm)",
    "(Corrected core speed) (rpm)",
    "(Bypass Ratio) ",
    "(Burner fuel-air ratio)",
    "(Bleed Enthalpy)",
    "(Required fan speed)",
    "(Required fan conversion speed)",
    "(High-pressure turbines Cool air flow)",
    "(Low-pressure turbines Cool air flow)"]

col_names = index_names + setting_names + sensor_names

df = pd.read_csv("../dataset/CMaps/train_FD001.txt",sep='\s+',header=None,index_col=False,names=col_names).sort_values(['id','time_cycles'])
print(df.shape)
df.sample(3)
```

```python
rul = pd.DataFrame(df.groupby('id')['time_cycles'].max()).reset_index()
rul.columns = ['id', 'max']
df = df.merge(rul, on=['id'], how='left')
df['RUL'] = df['max'] - df['time_cycles']
df.drop('max', axis=1, inplace=True)
df.sample(3)
```

```python
list_drop_col = []

for col in df.columns:
    if not np.std(df[col]):
        list_drop_col.append(col)

list_drop_col
```

```python
list_columns_droped = ['time_cycles','setting_3','(Fan inlet temperature) (◦R)','(Engine pressure ratio(P50/P2)',
                    '(Required fan speed)','(Required fan conversion speed)']

df.drop(columns=list_columns_droped, inplace=True)
print(df.shape)
df.sample(3)
```

```python
feature_columns = df.columns.tolist()
feature_columns.remove("RUL")
#feature_columns.remove("id")

flag_test = (df["id"] == 1) | (df["id"] == 10) | (df["id"] == 20) | (df["id"] == 30) | (df["id"] == 40) | \
            (df["id"] == 50) | (df["id"] == 60) | (df["id"] == 70) | (df["id"] == 80)| (df["id"] == 90)  

y = df["RUL"]
X = df[feature_columns]


X_train = X[~flag_test]
X_test = X[flag_test]
y_train = y[~flag_test]
y_test = y[flag_test]
```

```python
from sklearn.model_selection import RepeatedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
import xgboost as xgb


feature_columns = df.columns.tolist()
feature_columns.remove("RUL")
feature_columns.remove("id")

X = df[feature_columns]
y = df['RUL']
groups = df['id']

# Criar o regressor XGBoost
xg_reg = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)

# Definir o número de splits e repetições
n_splits = 10
n_repeats = 10

# Inicializar RepeatedKFold
rkf = RepeatedKFold(n_splits=n_splits, n_repeats=n_repeats, random_state=42)

# Realizar a validação cruzada repetida
mae_scores = []

for train_index, test_index in rkf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    

    # Normaliza
    scaler = StandardScaler()
    scaler.fit(X_train)

    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

    # Ajustar o modelo usando os dados de treino do fold atual
    xg_reg.fit(X_train, y_train)
    
    # Fazer previsões no conjunto de teste do fold atual
    y_pred = xg_reg.predict(X_test)
    
    # Calcular o MAE para o fold atual
    mae = mean_absolute_error(y_test, y_pred)
    mae_scores.append(mae)

# Converter a lista de resultados em um array NumPy
mae_scores = np.array(mae_scores)

# Imprimir os resultados
print(f"Erro Médio Absoluto (MAE) para cada repetição: {mae_scores}")
print(f"Média do MAE: {np.mean(mae_scores):.2f}")
print(f"Desvio padrão do MAE: {np.std(mae_scores):.2f}")
```
